## Définition :  
Le format CSV (pour **comma separated values**, soit en français valeurs séparées par des virgules) est un format très utilisé pour représenter des données structurées, notamment pour importer ou exporter des données à partir d'une feuille de calculs d'un tableur.  
C'est un fichier texte dans lequel chaque ligne correspond à une ligne du tableau




Exemple de feuille de calculs  
im1.png

**Fichier csv correspondant :**  
Nom,Prénom,NSI,Physique,Maths  
Baron,Fathi,18,16,15  
Blanc,Evan,2,5,1  
Boringer,Aline,16,13,14  
Cerdan,Baptiste,12,10,8  

Dans ce format, chaque ligne représente un enregistrement c'est à dire une structure de données, de types éventuellement différent auxquelles on accède grâce à un nom et, sur une même ligne, les différents champs de l’enregistrement sont réparés par une virgule (d’où le nom).

### Import d'un fichier csv

* La bibliothèque csv implémente des classes pour lire des données tabulaires au format CSV

* La fonction <code>reader()</code> du module csv renvoie un objet de type csv.reader qui est itérable. Chaque élément de cet objet est une liste.

* La fonction <code>DictReader</code> du module csv renvoie un objet de type csv.DictReader itérable aussi. Chaque élément de cet objet est une liste de tuple à 2 éléments. La première ligne du fichier csv se retrouve être les premiers éléments de chaque tuple.

# Utilisation de la fonction reader  
Copier le chemin d'accès de votre fichier csv puis coller cette adresse dans le code suivant : 


```python
import csv # le module pour les fichiers csv
file = open("/content/drive/MyDrive/1_NSI_2020_2021/Python/prof_eleves.csv" , "r") # ouvrir le fichier
csv_en_liste = csv.reader(file , delimiter = ",") # initialisation d’un lecteur de fichier delimiter est facultatif
for ligne in csv_en_liste : # parcours du lecteur avec une boucle
    print(ligne) # affichage ligne à ligne
file.close () # fermeture du fichier
```

On obtient des listes pour chacune des lignes.

# Utilisation de la fonction Dictreader

```python

import csv # le module pour les fichiers csv
file = open("/content/drive/MyDrive/1_NSI_2020_2021/Python/prof_eleves.csv" , "r") # ouvrir le fichier
csv_en_dico = csv.DictReader(file , delimiter = ",") # initialisation d’un lecteur de fichier avec création automatique de dictionnaire
for ligne in csv_en_dico : # parcours du lecteur avec une boucle
	print(dict(ligne)) # affichage ligne à ligne
file.close () # fermeture du fichier
```

On obtient chaque ligne sous forme de dictionnaire.
